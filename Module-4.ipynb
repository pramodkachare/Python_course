SYNTAX:
    plt.plot(y) 
    -> plots the given data on y-axis (vertical). The x-axis will denote the index values for each data point.
    
    plt.plot(x, y)
    -> Plots data points in y (vertical) according to index position in x (horizontal).
    
    plt.plot(x1, y1, x2, y2, x3, y3, ....)
    -> It plots the individual lines corresponding to (x1, y1), (x2, y2), (x3, y3),.... with different colors.

```python
t = np.arange(0, 1.01, 0.01)

x = 0.5*np.sin(2*np.pi*4*t)

plt.plot(t, x)
plt.xlabel('Time')      # <- Label on x-axis (horizontal)
plt.ylabel('Aplitude')  # <- Label on y-axis (vertical)
plt.ylim([-1, 1])       # Fix Lower and Upper limit for y-axis (vertical) display
plt.xlim([0, 1])       # Fix Lower and Upper limit for x-axis (horizontal) display
plt.grid('minor')
plt.title('Sine wave')
plt.savefig('sine.png')
plt.show()
```

```python
t = np.arange(0, 1.01, 0.01)

x1 = 0.5*np.sin(2*np.pi*4*t)
x2 = 0.5*np.cos(2*np.pi*4*t)

plt.plot(t, x1, t, x2)
plt.xlabel('Time')      # <- Label on x-axis (horizontal)
plt.ylabel('Aplitude')  # <- Label on y-axis (vertical)
plt.ylim([-1, 1])       # Fix Lower and Upper limit for y-axis (vertical) display
plt.xlim([0, 1])        # Fix Lower and Upper limit for x-axis (horizontal) display
plt.grid('minor')
plt.legend(['Sine', 'Cosine']) # <- adds information of plot with name (NOTE: Names must be given order.)
# plt.title('Sine wave')
# plt.savefig('sine.png')
plt.show()
```

```python
t = np.arange(0, 1.01, 0.01)

x1 = 0.5*np.sin(2*np.pi*4*t)
x2 = 0.5*np.cos(2*np.pi*4*t)

tt = np.arange(0, 0.51, 0.01)
y = tt**2

plt.figure(figsize=(10, 5))  # ,_ generate figure of (width, height)
plt.plot(t, x1, t, x2, tt, y)
plt.xlabel('Time')      # <- Label on x-axis (horizontal)
plt.ylabel('Aplitude')  # <- Label on y-axis (vertical)
plt.ylim([-1, 1])       # Fix Lower and Upper limit for y-axis (vertical) display
plt.xlim([0, 1])        # Fix Lower and Upper limit for x-axis (horizontal) display
plt.grid('minor')
plt.legend(['Sine', 'Cosine', 'Square']) # <- adds information of plot with name (NOTE: Names must be given order.)
# plt.title('Sine wave')
plt.savefig('sine.png')
plt.show()
```

```python
SYNTAX:
    plt.plot(x, y, '-r')  or ax[i, j].plot(x, y, '-r')
    
    Coding of styles:  {marker_style}{line_style}{color}
    -> It plots the curve with x on horizontal axis and y on vertical axis with continuous line (-) having red color (r).
    -> Other possible colors:
            r -> red
            c -> cyan
            k -> black
            m -> magenta
            y -> yelloow
            g -> green
            b -> blue (default)
            o -> orange
    -> Other possible line styles:
            - -> Continuous line (default: only when marker is also not mentioned)
            -- -> Dashed line
            .. -> Dotted line
            .- -> Dot-dash line (continuous line with dots at (x, y) positions)
    
```

```python
fig, ax = plt.subplots(2, 2, constrained_layout=True)

ax[0, 0].plot(t, x1, t, x2)
ax[0, 0].set_xlabel('Time')
ax[0, 0].set_ylabel('Sine')
ax[0, 0].set_title('Sine wave')
ax[0, 0].grid(True)

ax[1, 1].plot(t, x2, '*-g')
ax[1, 1].set_frame_on(False)
ax[1, 1].set_xlabel('Time')
ax[1, 1].set_ylabel('Cosine')
ax[1, 1].set_title('Cosine wave')
```

### Types of plots

```python
Custom colors:
    [Red, Green, Blue] = each of these are in range (0, 1)
    red color -> [1, 0, 0]
    green color -> [0, 1, 0]
    blue color -> [0, 0, 1]
    white color -> [1, 1, 1]
    black color -> [0, 0, 0]
    
    alpha = coefficient of transparency (0-1, 0: completely transparent, 1: Opaque)
```

```python
marks = [95, 70, 65, 83, 79]
names = ['A', 'B', 'C', 'D', 'E']

fig, ax = plt.subplots(1, 2, constrained_layout=True)
ax[0].bar(names, marks, width = 1.4, facecolor = [0.5, 0.5, 0], edgecolor = 'r', alpha = 0.5)
ax[0].set_yticks([0, 50, 100])
ax[0].set_yticklabels(['L', 'M', 'H'])

ax[1].barh(np.arange(5), marks, height = 0.95)
```

## Stem

```python
x = np.arange(10)
y = x**2

plt.plot(x, y)
plt.stem(x, y, 'r', linefmt='--')
plt.show()
```

## Imshow

```python
x = np.random.rand(50, 50)

plt.imshow(x, origin='lower')
plt.colorbar()
plt.set_cmap('jet')
plt.show()
```

## Pcolormesh

```python
x = np.arange(10)
y = [1, 2, 3]

x_m, y_m = np.meshgrid(x, y)

Z = x_m + y_m

plt.pcolormesh(x_m, y_m, Z)
plt.colorbar()
plt.set_cmap('jet')
plt.show()
```

```python
fig, ax = plt.subplots()

t = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(t, np.sin(t))

def update(i):
    line.set_xdata(t+i/100)
    ax.set_title(i)
    return line

ani = FuncAnimation(fig, update, interval = 100, blit=False, repeat=True)

plt.show()
```
